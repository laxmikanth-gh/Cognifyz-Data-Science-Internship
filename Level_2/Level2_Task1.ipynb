```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset/Dataset .csv")

print(df.head())
```

```python
table_booking = df["Has Table booking"].value_counts()

percentage_table_booking = (
    table_booking["Yes"] / len(df)
) * 100

print(f"Percentage of restaurants offering table booking: {percentage_table_booking:.2f}%")
```

```python
online_delivery = df["Has Online delivery"].value_counts()

percentage_online_delivery = (
    online_delivery["Yes"] / len(df)
) * 100

print(f"Percentage of restaurants offering online delivery: {percentage_online_delivery:.2f}%")
```

```python
table_rating = df.groupby("Has Table booking")["Aggregate rating"].mean()

print(table_rating)
```

```python
delivery_price = pd.crosstab(
    df["Price range"],
    df["Has Online delivery"]
)

print(delivery_price)
```

```python
table_booking.plot(kind="bar")
plt.title("Table Booking Availability")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
online_delivery.plot(kind="bar")
plt.title("Online Delivery Availability")
plt.ylabel("Number of Restaurants")
plt.show()
```
